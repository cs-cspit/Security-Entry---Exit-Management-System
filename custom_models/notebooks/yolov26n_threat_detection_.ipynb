```python
from ultralytics import YOLO
import os

# Create a basic data.yaml file for your dataset with corrected paths
data_yaml_content = """
train: /content/dataset/combined_gunsnknifes/train/images
val: /content/dataset/combined_gunsnknifes/val/images
nc: 1
names: ['human']
"""

# Ensure the directory for data.yaml exists (if not created by a previous cell)
os.makedirs('/content/dataset', exist_ok=True)

# Save it to the correct path, overwriting if it exists
with open('/content/dataset/data.yaml', 'w') as f:
    f.write(data_yaml_content)

print("data.yaml file updated with correct paths!")

# Initialize the YOLOv26 model with the pre-trained weights
model = YOLO("yolo26n.pt")  # Using yolov26n.pt as the base weights

# Train the model on your custom dataset
results = model.train(
    data="/content/dataset/data.yaml",  # Path to your dataset YAML file
    epochs=50,                         # Number of epochs
    imgsz=640,                         # Image size (640x640)
    batch=16,                          # Batch size
    device=0,                          # Use GPU (device 0)
    project="runs/yolo26_threat_detection_for_room_camera",        # Project folder to save results
    name="exp2"                        # Experiment name
)

# Optionally, display the results (after training)
print(results)
```
