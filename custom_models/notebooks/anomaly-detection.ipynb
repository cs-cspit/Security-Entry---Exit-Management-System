```python
import cv2
import os

video_root = "/kaggle/input/datasets/webadvisor/real-time-anomaly-detection-in-cctv-surveillance/data"
output_root = "/kaggle/working/frames"

selected_classes = ["normal", "robbery", "fighting", "arson", "explosion", "burglary", "explosion", "shooting", "vandalism", "stealing", "assault"]

FRAME_SKIP = 10
MAX_FRAMES = 100
MAX_VIDEOS = 100

# os.makedirs(output_root, exist_ok=True)

# for cls in selected_classes:
#     cls_path = os.path.join(video_root, cls)
#     out_cls = os.path.join(output_root, cls)
#     os.makedirs(out_cls, exist_ok=True)

#     videos = os.listdir(cls_path)[:MAX_VIDEOS]

#     for video in videos:
#         video_path = os.path.join(cls_path, video)

#         cap = cv2.VideoCapture(video_path)
#         count = 0
#         saved = 0

#         while True:
#             ret, frame = cap.read()
#             if not ret or saved >= MAX_FRAMES:
#                 break

#             if count % FRAME_SKIP == 0:
#                 filename = f"{video[:-4]}_{saved}.jpg"
#                 cv2.imwrite(os.path.join(out_cls, filename), frame)
#                 saved += 1

#             count += 1

#         cap.release()

# print("✅ Optimized extraction done")

os.makedirs(output_root, exist_ok=True)

for cls in selected_classes:
    cls_path = os.path.join(video_root, cls)
    out_cls = os.path.join(output_root, cls)
    os.makedirs(out_cls, exist_ok=True)

    videos = os.listdir(cls_path)[:MAX_VIDEOS]

    for video in videos:
        video_path = os.path.join(cls_path, video)

        if not video.endswith(".mp4"):
            continue

        cap = cv2.VideoCapture(video_path)
        count, saved = 0, 0

        while True:
            ret, frame = cap.read()
            if not ret or saved >= MAX_FRAMES:
                break

            if count % FRAME_SKIP == 0:
                filename = f"{video[:-4]}_{saved}.jpg"
                cv2.imwrite(os.path.join(out_cls, filename), frame)
                saved += 1

            count += 1

        cap.release()

print("✅ Frames created")
```

```python
import shutil
from sklearn.model_selection import train_test_split

base = "/kaggle/working/frames"
output = "/kaggle/working/yolo_dataset"

classes = os.listdir(base)

for split in ["train", "val"]:
    for cls in classes:
        os.makedirs(f"{output}/{split}/{cls}", exist_ok=True)

for cls in classes:
    cls_path = os.path.join(base, cls)
    images = os.listdir(cls_path)

    train, val = train_test_split(images, test_size=0.2, random_state=42)

    for img in train:
        shutil.copy(os.path.join(cls_path, img),
                    f"{output}/train/{cls}/{img}")

    for img in val:
        shutil.copy(os.path.join(cls_path, img),
                    f"{output}/val/{cls}/{img}")

print("✅ Dataset ready")
```
