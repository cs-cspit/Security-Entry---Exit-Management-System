```python
import kagglehub
import os

# Dataset handle from the URL you provided
handle = "webadvisor/real-time-anomaly-detection-in-cctv-surveillance"

# Your selected categories
categories = [
    "abuse", "arson", "assault", "burgurly", "fighting",
    "normal", "robbery", "stealing", "shooting", "vandalism"
]

# Dictionary to store the local paths of each folder
paths = {}

for category in categories:
    print(f"Downloading {category}...")
    # 'path' argument targets the specific subfolder in the Kaggle structure
    paths[category] = kagglehub.dataset_download(handle, path=f"data/{category}")

print("\nAll downloads complete. Sample path:")
print(paths["fighting"])
```
